## Comparitive Analysis of the teams

```python
df.columns
```

```python
df[['team1','team2']]
```

```python
total_matches_played = (df.team1.value_counts() + df.team2.value_counts()).to_frame().reset_index()
```

```python
total_matches_played.columns = ['team_name','matches_played']
```

```python
total_matches_played
```

```python
matches_won = df.winner.value_counts().reset_index()
matches_won.columns = ['team_name','matches_won']
matches_won
```

```python
played = total_matches_played.merge(matches_won,on='team_name',how='left')
played
```

```python
trace1 = go.Bar(x=played.team_name, y=played.matches_played, name = 'Total Matches')
trace2 = go.Bar(x=played.team_name, y=played.matches_won, name = 'Matches Won')
```

```python
data = [trace1,trace2]
```

```python
iplot(data)
```
